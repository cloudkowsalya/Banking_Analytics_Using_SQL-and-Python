# 1) Print customer Id, customer name and average account_balance maintained by each customer for all  of his/her accounts in the bank.(8 Rows)

```python
query = ('''SELECT
    c.customer_id,
    c.customer_name,
    AVG(a.Balance_Amount) AS avg_balance
FROM CUSTOMER_INFORMATION c
JOIN ACCOUNT_DATA a
    ON c.customer_id = a.Customer_id
GROUP BY c.customer_id, c.customer_name;
''')
result = pd.read_sql_query(query,conn)
headers = [desc[0] for desc in cursor.description]
print(tabulate(result,headers=headers,tablefmt='grid'))
```

# 2.Print customer_id , account_number and balance_amount , condition that if balance_amount is nil then assign transaction_amount  for account_type = "Credit Card"(4 Rows)

```python
query =('''
    select
    m.Customer_id,
    d.Account_Number,
    COALESCE(d.Balance_Amount, t.Transaction_amount) AS final_balance
FROM ACCOUNT_DATA d
JOIN ACCOUNT_MAPPING m
    ON d.Account_Number = m.Account_Number
LEFT JOIN TRANSACTION_LOG t
    ON d.Account_Number = t.Account_Number
WHERE d.Account_Type = 'Credit Card'
LIMIT 4;''')
result = pd.read_sql_query(query,conn)

print(tabulate(result,headers='keys',tablefmt='grid'))
```

# 3.Print account_number and balance_amount , transaction_amount,Transaction_Date from Bank_Account_Details and bank_account_transaction for all the transactions occurred during march,2020 and april, 2020(12 Rows)

```python
query ='''
SELECT
    d.Account_Number,
    d.Balance_Amount,
    t.Transaction_amount,
    t.Transaction_Date
FROM ACCOUNT_DATA d
JOIN TRANSACTION_LOG t
    ON d.Account_Number = t.Account_Number
WHERE strftime('%m', t.Transaction_Date) IN ('03', '04')
  AND strftime('%Y', t.Transaction_Date) = '2020'
ORDER BY t.Transaction_Date;'''

result = pd.read_sql_query(query,conn)
print(tabulate(result,headers='keys',tablefmt='grid'))
```

# 4.Print all of the customer id, account number,  balance_amount, transaction_amount , Transaction_Date  from bank_customer, Bank_Account_Details and bank_account_transaction tables where excluding all of their transactions in march, 2020  month (22 Rows)

```python
query='''
SELECT
    c.Customer_ID,
    d.Account_Number,
    d.Balance_Amount,
    t.Transaction_Amount,
    t.Transaction_Date
FROM CUSTOMER_INFORMATION c
JOIN ACCOUNT_DATA d
    ON c.Customer_ID = d.Customer_ID
JOIN TRANSACTION_LOG t
    ON d.Account_Number = t.Account_Number
WHERE NOT (
    strftime('%m', t.Transaction_Date) = '03'
    AND strftime('%Y', t.Transaction_Date) = '2020'
)
ORDER BY c.Customer_ID, t.Transaction_Date;'''

result = pd.read_sql_query(query,conn)
print(tabulate(result,headers='keys',tablefmt='grid'))

```

#5.Print only the customer id, account_number, balance_amount,transaction_amount ,transaction_date who did transactions during the first quarter. Do not display the accounts if they have not done any transactions in the first quarter.(16 Rows)

```python
query='''
SELECT
    c.Customer_ID,
    d.Account_Number,
    d.Balance_Amount,
    t.Transaction_Amount,
    t.Transaction_Date
FROM CUSTOMER_INFORMATION c
JOIN ACCOUNT_DATA d
    ON c.Customer_ID = d.Customer_ID
JOIN TRANSACTION_LOG t
    ON d.Account_Number = t.Account_Number
WHERE strftime('%Y', t.Transaction_Date) = '2020'
  AND strftime('%m', t.Transaction_Date) IN ('01','02','03')
ORDER BY c.Customer_ID, t.Transaction_Date;'''

result = pd.read_sql_query(query,conn)
print(tabulate(result,headers='keys',tablefmt='grid'))
```

#7.Print all Customer_id, Account_Number, Account_type, and display deducted balance_amount by  subtracting only negative transaction_amounts for Relationship_type ="P" ( P - means  Primary , S - means Secondary ) .(27 Rows)

```python
query='''
SELECT
    m.Customer_ID,
    d.Account_Number,
    d.Account_Type,
    d.Balance_Amount + COALESCE(SUM(
        CASE
            WHEN t.Transaction_Amount < 0 THEN t.Transaction_Amount
            ELSE 0
        END
    ), 0) AS Deducted_Balance
FROM ACCOUNT_MAPPING m
JOIN ACCOUNT_DATA d
    ON m.Account_Number = d.Account_Number
LEFT JOIN TRANSACTION_LOG t
    ON d.Account_Number = t.Account_Number
WHERE d.Relationship_Type = 'P'
GROUP BY m.Customer_ID, d.Account_Number, d.Account_Type, d.Balance_Amount
ORDER BY m.Customer_ID;'''


cursor.execute(query)
rows = cursor.fetchall()
headers = [desc[0] for desc in cursor.description]

print(tabulate(rows, headers=headers, tablefmt="grid"))
```

#9 c) After retrieving all records of accounts and their linked accounts, display the   transaction amount of accounts appeared  in another column.(26 Rows)

```python
query = '''
SELECT
    a.Account_Number,
    m.Linking_Account_Number,
    a.Account_Type,
    a.Balance_Amount,
    t.Transaction_Amount AS Account_Transaction,
    t2.Transaction_Amount AS Linked_Account_Transaction
FROM ACCOUNT_DATA a
JOIN ACCOUNT_MAPPING m
    ON a.Account_Number = m.Account_Number
LEFT JOIN TRANSACTION_LOG t
    ON a.Account_Number = t.Account_Number
LEFT JOIN TRANSACTION_LOG t2
    ON m.Linking_Account_Number = t2.Account_Number
ORDER BY a.Account_Number;
'''

cursor.execute(query)
rows = cursor.fetchall()
headers = [description[0] for description in cursor.description]

print(tabulate(rows, headers=headers, tablefmt="grid"))
```

#10) Display all saving account holders have “Add-on Credit Cards" and “Credit cards" (3 Rows)

```python
query = '''
SELECT DISTINCT
    c.Customer_ID,
    c.Customer_Name,
    s.Account_Number AS Savings_Account,
    cc.Account_Number AS Credit_Card_Account,
    ac.Account_Number AS AddOn_Credit_Card_Account
FROM CUSTOMER_INFORMATION c
JOIN ACCOUNT_DATA s
    ON c.Customer_ID = s.Customer_ID AND s.Account_Type = 'SAVINGS'
JOIN ACCOUNT_DATA cc
    ON c.Customer_ID = cc.Customer_ID AND cc.Account_Type = 'CREDIT CARD'
JOIN ACCOUNT_DATA ac
    ON c.Customer_ID = ac.Customer_ID AND ac.Account_Type = 'ADD-ON CREDIT CARD'
ORDER BY c.Customer_ID;
'''

cursor.execute(query)
rows = cursor.fetchall()
headers = [description[0] for description in cursor.description]

print(tabulate(rows, headers=headers, tablefmt="grid"))
```
